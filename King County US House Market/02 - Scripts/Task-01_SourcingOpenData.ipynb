```python
# Define a function to reorganize and rename Waterfront Column

def waterfront_text(row):

  if row['waterfront'] == 0:
    return 'No'
  elif row['waterfront'] == 1:
    return 'Yes'
  else: return 'n/a'
```

```python
# Create text label for each row based on waterfront field

df_house['waterfront_text'] = df_house.apply(waterfront_text, axis=1)
```

```python
# Define a function to reorganize and rename "View" Column

def view_text(row):

  if row['view'] == 0:
    return 'No View'
  elif row['view'] == 1:
    return 'Fair'
  elif row['view'] == 2:
    return 'Average'
  elif row['view'] == 3:
    return 'Good'
  elif row['view'] == 4:
    return 'Excellent'
  else: return 'n/a'
```

```python
# Create a text label for each row based on the view column

df_house['view_text'] = df_house.apply(view_text, axis=1)
```

```python
# Define a function to reorganize and rename "Condition" Column

def house_condition(row):

  if row['condition'] == 1:
    return 'Poor'
  elif row['condition'] == 2:
    return 'Fair'
  elif row['condition'] == 3:
    return 'Average'
  elif row['condition'] == 4:
    return 'Good'
  elif row['condition'] == 5:
    return 'Very Good'
  else: return 'n/a'
```

```python
# Generate a condition label for each house using the house_condition function

df_house['house_condition'] = df_house.apply(house_condition, axis=1)
```

```python
# Inspecting Dataframe

df_house.head(10)
```

```python
# Define a function to reorganize and rename "Grade" Column

def grade_text(row):

  if row['grade'] == 1:
    return 'Cabin'
  if row['grade'] == 2:
    return 'Substandard'
  if row['grade'] == 3:
    return 'Poor'
  if row['grade'] == 4:
    return 'Low'
  if row['grade'] == 5:
    return 'Fair'
  if row['grade'] == 6:
    return 'Low Average'
  if row['grade'] == 7:
    return 'Average'
  if row['grade'] == 8:
    return 'Good'
  if row['grade'] == 9:
    return 'Better'
  if row['grade'] == 10:
    return 'Very Good'
  if row['grade'] == 11:
    return 'Excellent'
  if row['grade'] == 12:
    return 'Luxury'
  if row['grade'] == 13:
    return 'Mansion'
  else: return 'n/a'
```

```python
# Create a text label for each row based on the grade value

df_house['grade_text'] = df_house.apply(grade_text, axis=1)
```
